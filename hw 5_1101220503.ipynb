```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
```

```python
df = pd.read_csv("my2020-fuel-consumption-ratings.csv")
# melihat dataset
df.head()
```

```python
# merangkum data
df.describe()
```

```python
df[['Model year','Engine size (L)','Cylinders','CO2 rating']]
df.head(10)
```

```python
df[['Cylinders','Engine size (L)','Model year','CO2 rating']]
df.hist()
plt.show()
```

```python
plt.scatter(df['Model year'], df['Engine size (L)'], color='blue')
plt.xlabel("Model year")
plt.ylabel("Engine size (L)")
plt.show()

plt.scatter(df['Engine size (L)'], df['CO2 rating'], color='blue')
plt.xlabel("Engine size (L)")
plt.ylabel("CO2 rating")
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
plt.scatter(train['Engine size (L)'], train['CO2 rating'], color='blue')
plt.xlabel("Engine size (L)")
plt.ylabel("CO2 rating")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Engine size (L)']])
train_y = np.asanyarray(train[['CO2 rating']])
regr.fit (train_x, train_y)
```

```python
print('Coefficients: ', regr.coef_)
print('Intercept: ', regr.intercept_)
```

```python
plt.scatter(train['Engine size (L)'], train['CO2 rating'], color='blue')

# Plot the regression line
plt.plot(train['Engine size (L)'], regr.coef_[0] * train['Engine size (L)'] + regr.intercept_, '-r')

plt.xlabel("Engine size (L)")
plt.ylabel("CO2 rating")
plt.show()
```

```python
from sklearn.metrics import r2_score
import numpy as np

test_x = np.asarray(test[['Engine size (L)']])  # Correct variable name
test_y = np.asarray(test[['CO2 rating']])

test_y_pred = regr.predict(test_x)  # Prediction using the model

# Calculate the metrics
print("Mean absolute error: %.2f" % np.mean(np.abs(test_y_pred - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_pred - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y, test_y_pred))  # Correct R2 calculation
```
